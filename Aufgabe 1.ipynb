```python
import numpy as np
from pylab import * 
import matplotlib.pyplot as plt
```

```python
def conv(x):
    return x.replace(",", ".").encode()
%matplotlib inline
```

```python
#import data
A05_S10_V4_Aphoch = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A0.5_S10_V4_Aphoch.CSV")), delimiter =";")
A05_S10_V4_Apm = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A0.5_S10_V4_Apm.CSV")), delimiter =";")
A4_S1_V4_Apm = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S1_V4_Apm.CSV")), delimiter =";")
A4_S10_V3_Apm = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S10_V3_Apm.CSV")), delimiter =";")
A4_S10_V4_Apm_Boxcar = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S10_V4_Apm_Boxcar.CSV")), delimiter =";")
A4_S10_V4_Apm_cosine = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S10_V4_Apm_cosine.CSV")), delimiter =";")
A4_S10_V4_Apm_dreieck = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S10_V4_Apm_dreieck.CSV")), delimiter =";")
A4_S10_V4_Apm = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S10_V4_Apm.CSV")), delimiter =";")
A4_S10_V4_Apo = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S10_V4_Apo.CSV")), delimiter =";")
A4_S10_V32_Apm = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S10_V32_Apm.CSV")), delimiter =";")
A4_S50_V4_Apm = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A4_S50_V4_Apm.CSV")), delimiter =";")
A32_S10_V4_Apm = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A32_S10_V4_Apm.CSV")), delimiter =";")
A32_S10_V4_Apo = np.genfromtxt((conv(x) for x in open("Aufgabe 1/CVS/A32_S10_V4_Apo.CSV")), delimiter =";")
```
